### Imputación de datos
+ Dato perdido: vacío en las DB
+ Causas: encuesta incompleta, abandono
+ Soluciones: reemplazar por un valor adecuado (imputación), eliminar datos (o columnas) asociados, generar dummy
+ Imputación individual o múltiple
+ Individual: media/moda, aleatorio desde columna, regresión
+ Múltiple: regresión iterativa cruzada, distribución de probabilidad conjunta
